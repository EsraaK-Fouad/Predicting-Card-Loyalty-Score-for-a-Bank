```python
historical_transactions['purchase_date'] =pd.to_datetime(historical_transactions['purchase_date'])
new_merchant_transactions['purchase_date'] =pd.to_datetime(new_merchant_transactions['purchase_date'])
diff_month_lag = historical_transactions['purchase_date'].max().month - new_merchant_transactions['purchase_date'].max().month
```

# Data Preprocessing 

- drop state_id ,subsector_id ,  merchant_category_id, merchant_category_id
- Drop rows that have null values and outliers 
- normalize skewed features 

```python
rows_index_to_drop = list(set(upper_limit_installments) | set(upper_limit_purchase_amount) |set(lower_limit_purchase_amount)| set(lower_limit_installments))
print("Total number of outliers " , len(rows_index_to_drop))
transactions_data.drop(index=rows_index_to_drop, inplace=True)
print("Shape of data after dropping outliers" , transactions_data.shape)
```

```python
len(rows_index_to_drop)
```

```python
transactions_data.isnull().any(axis=1).sum()
```

```python
transactions_data_cleaned = transactions_data.dropna(axis=0)
```

```python
print("Shape of data after dropping outliers and null values :" , transactions_data_cleaned.shape)
print("Number of missing values in transactions data")
transactions_data_cleaned.isnull().any(axis=1).sum()
```

```python
transactions_data_cleaned['purchase_amount'] = np.log(transactions_data_cleaned['purchase_amount'] + 1)
sns.distplot(transactions_data_cleaned['purchase_amount'])
```
